```python
import string

text = "Hello World! NLP is FUN üòÑ, isn't it?"
print("Texte original :", text)

# Tout en minuscules
text = text.lower()

# Supprimer la ponctuation
text = text.translate(str.maketrans('', '', string.punctuation))

print("Texte nettoy√© :", text)
```

```python
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords

text = "NLP is a fun and exciting field of Artificial Intelligence."

# Tokenisation
tokens = word_tokenize(text.lower())
print("Tokens :", tokens)

# Stopwords
stop_words = set(stopwords.words('english'))

# Filtrer les tokens
filtered_tokens = [word for word in tokens if word not in stop_words]
print("Tokens filtr√©s :", filtered_tokens)

# Ajouter un mot inutile √† la liste
stop_words.add('nlp')
filtered_tokens_custom = [word for word in tokens if word not in stop_words]
print("Tokens filtr√©s personnalis√©s :", filtered_tokens_custom)
```

```python
from nltk.stem import WordNetLemmatizer

lemmatizer = WordNetLemmatizer()

words = ["running", "flies", "cats", "better"]
lemmatized = [lemmatizer.lemmatize(w) for w in words]

print("Avant :", words)
print("Apr√®s :", lemmatized)
```

```python
import spacy
nlp = spacy.load("en_core_web_sm")

doc = nlp("The children are playing in the gardens")
lemmatized = [token.lemma_ for token in doc]
print("Lemmes :", lemmatized)
```
